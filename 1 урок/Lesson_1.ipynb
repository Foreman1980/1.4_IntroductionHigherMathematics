```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5, 5, 101)
x
```

```python
fig1 = pd.Series(np.cos(x), index=x)
fig2 = pd.Series(np.cos(3 * x), index=x)
```

```python
fig1.plot()
fig2.plot()
plt.axis([-5, 5, -1.1, 1.1])
plt.show()
```
